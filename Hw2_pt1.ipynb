```python
import math
import pandas as pd

data = pd.read_csv('gain_practice.csv')
print(data.head())

print("\nPart 1 Answers:\n")

#School Canceled Entropy
entropySchoolCanceled = -((3/5) * math.log2(3/5) + (2/5) * math.log2(2/5))

print("Pt1 Q2: Entropy of SC:", round(entropySchoolCanceled, 2))

#Ice Info Gain
entIceGivenSchool = -((1/3) * math.log2(1/3) + 2/3 * math.log2(2/3))
print("\nEntropy of Ice Cream given School Canceled:", round(entIceGivenSchool, 2))

gainIce = entropySchoolCanceled - (2/5 * 0 + 3/5 * entIceGivenSchool) 
print("\nPt1 Q3: Information Gain of Ice:", round(gainIce, 2))

#Above Freezing Info Gain
entAboveFYes = -((2/3) * math.log2(2/3) + (1/3) * math.log2(1/3))
entAboveFNo = -((1/2) * math.log2(1/2) + (1/2) * math.log2(1/2))
print("\nentAboveFNo:", round(entAboveFNo, 2))
print("entAboveFYes:", round(entAboveFYes, 2))

gainAboveFreezing = entropySchoolCanceled - (2/5 * entAboveFNo + 3/5 * entAboveFYes)
print("\nPt1 Q4: Information Gain of Above Freezing:", round(gainAboveFreezing, 2))

print("\nPt1 Q5: Ice has higher information gain, so we would split on that attribute first.")
```
